```python
import pandas as pd
import xlrd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import csv
```

```python
df = pd.read_excel('parced_courses.xlsx')
```

```python
df.tail(60)
```

```python
def rating(x):
    return x[:3]
```

```python
df['Rating'] = df['Rating'].apply(rating)
```

```python
df = df.loc[df['Price'] != '0 ₽']
```

```python
df
```

```python
def start(x):
    if x == 'Старт в любое время':
        return 1
    else:
        return 0
```

```python
df['Start'] = df['Start'].apply(start)
```

```python
df['Start'].value_counts()
```

```python
df['School'].value_counts()
```

```python
df = df.drop('School', axis = 1)
```

```python
df[['Price_with_sale', 'Price_without_sale']] = df['Price'].str.split('₽', 1, expand = True)
```

```python
df
```

```python
df.dtypes
```

```python
df = df.drop('Price', axis = 1)
```

```python
def sep_0s(x):
    k = str(x).split()
    return ''.join(k[:-1])
```

```python
def sep_1s(x):
    k = str(x).split()
    return ''.join(k)
```

```python
df['Price_without_sale'] = df['Price_without_sale'].apply(sep_0s)
```

```python
df['Price_with_sale'] = df['Price_with_sale'].apply(sep_1s)
```

```python
df['Price_with_sale'] = df['Price_with_sale'].astype('int')
```

```python
df['Price_without_sale'] = df['Price_without_sale'].astype('int')
```

```python
df.dtypes
```

```python
df['Sale'] = round((df['Price_without_sale'] - df['Price_with_sale'])/df['Price_with_sale']*100, 0)
```

```python
ds = df[['Price_with_sale', 'Price_without_sale', 'Sale', 'Duration', 'Rating',
         'Teacher', 'Chat', 'Lifetime_access', 'Certificate', 'Start',
        'Installment']]
```

```python
ds = ds.dropna()
```

```python
def duration(x):
    lst = x.split()
    return lst[0]
```

```python
ds['Duration'] = ds['Duration'].apply(duration)
```

```python
d = dd.drop('Price_without_sale', axis = 1)
```

```python
d1 = d.rename(columns = {'Price_with_sale': 'Price'})
```

```python
def picture_hist(parameter, name_x, name_y):
    plt.figure(figsize = (6, 6))
    plt.hist(d1[parameter], bins = 40, ec = 'grey', color = 'pink')
    plt.xlabel(name_x)
    plt.ylabel(name_y)
    plt.show()
```

```python
picture_hist('Price', 'Price', 'Number of courses')
```

```python
picture_hist('Duration', 'Duration', 'Number of courses')
```

```python
def picture_new_log(parameter, name_x, name_y):
    plt.figure(figsize = (6, 6))
    plt.hist(np.log(d1[parameter]), bins = 15, ec = 'grey', color = '#f5f5dc')
    plt.xlabel(name_x)
    plt.ylabel(name_y)
    plt.show()
```

```python
picture_new_log('Price', 'log(Price)', 'Number of courses')
```

```python
picture_new_log('Duration', 'log(Duration)', 'Number of courses')
```
